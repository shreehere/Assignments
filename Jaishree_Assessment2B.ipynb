###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
Exceeded_expectations=df.loc[(df['CodeKata Score']>15000)]
Reached_expectations=df.loc[(df['CodeKata Score']>10000) & (df['CodeKata Score']<15000)]
Needs_Improvement=df.loc[(df['CodeKata Score']>7000) & (df['CodeKata Score']<10000)]
Unsatisfactory=df.loc[(df['CodeKata Score']<7000)]
```

```python
#Downloading all the csv files with the given conditions

Exceeded_expectations.to_csv('Exceeded_expectations.csv')
files.download('Exceeded_expectations.csv')

Reached_expectations.to_csv('Reached_expectations.csv')
files.download('Reached_expectations.csv')

Needs_Improvement.to_csv('Needs_Improvement.csv')
files.download('Needs_Improvement.csv')

Unsatisfactory.to_csv('Unsatisfactory.csv')
files.download('Unsatisfactory.csv')
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

```python
#Calculating Average for each and every row in df
df["AVERAGE"]=((df['CodeKata Score']-df['Previous Geekions'])/(df['Previous Geekions']))*100
print(df)
print("-------------------------------------------")

#Total Average is calculated for the calculated average in df
Average=df["AVERAGE"].sum()/len(df)
print("Average of previous week geekions vs this week geekoins is :", Average)
```

###### No of students participated 

```python
students=df['Name'].count()
print("Number of students participated :", students)
```

###### #Average completion of python course or my_sql or python english or computational thinking

```python
#Replacing Nan values to zero for the average calculation
df.fillna(0,inplace=True)
```

```python
#Average completion of python course
avg_python=(len(df[df['python']==100])/len(df))*100

#Average completion of my_sql course
avg_mysql=(len(df[df['mysql']==100])/len(df))*100

#Average completion of python english
avg_pythonen=(len(df[df['python_en']==100])/len(df))*100

#Average completion of computational thinking
avg_ct=(len(df[df['computational_thinking']==100])/len(df))*100


print("Average completion of python course :", avg_python)
print("Average completion of my_sql course :", avg_mysql)
print("Average completion of python_en course :", avg_pythonen)
print("Average completion of computational_thinking course :", avg_ct)
print("---------------------------------------------------------------")

#DataFrame created with either python or mysql or python_en or computationalthinking is fully completed

df_avg_completion=df[(df['python']==100)|(df['mysql']==100)|(df['python_en']==100)|(df['computational_thinking']==100)]
df_avg_completion

Tot_avg=(len(df_avg_completion)/len(df))*100

print("Average completion with either pf python OR mysql OR python_en OR computational thinking")
print("----------------------------------------------------------------------------------------")
print("*************", Tot_avg ,"***************")
```

###### rising star of the week (top 3 candidate who performed well in that particular week)

```python
#sorting according to rising and taking the top 3 values 
rising_stars=df.sort_values(by='Rising',ascending=False).head(3)
rising_stars
```

```python
fig = px.histogram(rising_stars, x="Name",y="Rising",nbins=1000,title="******Rising stars of the week*****")
fig.show()
```

###### Shining stars of the week (top 3 candidates who has highest geekions)

```python
shining_stars=df.sort_values(by='Previous Geekions',ascending=False).head(3)
shining_stars
```

```python
fig1 = px.histogram(shining_stars, x="Name",y="Previous Geekions",nbins=10000,title="***************Shining stars of the week**************")
fig1.show()
```

###### Department wise codekata performence (pie chart)

```python
plt.style.use("fivethirtyeight")
fig=plt.figure(figsize=(5,5))
df.groupby(by="Department").sum()["CodeKata Score"].plot.pie(autopct="%.2f%%",fontsize=15)
plt.legend(loc=(1,0.7),fontsize=15)
plt.title("***** Department wise Codekata Performance *****")
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)
